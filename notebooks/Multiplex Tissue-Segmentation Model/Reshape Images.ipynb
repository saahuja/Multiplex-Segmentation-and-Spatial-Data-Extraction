# Manually Cropping (not ideal)

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np
image = mpimg.imread("DAPI1.jpg")
```

```python
image = image[:,...,0]
plt.imshow(image)
```

```python
image.shape
```

```python
from PIL import Image 
image = Image.fromarray(image)
```

```python
left = 400
right = 1344
top = 60
bottom = 1004
image = image.crop((left, top, right, bottom))
```

```python
image.show()
```

```python
image = np.array(image)
```

```python
plt.imshow(image)
```

# Resize 

```python
image = mpimg.imread("DAPI1.jpg")
```

```python
image = image[:,...,0]
plt.imshow(image)
```

```python
from PIL import Image 
image = Image.fromarray(image)
```

```python
sqrWidth = np.ceil(np.sqrt(image.size[0]*image.size[1])).astype(int)
im_resize = image.resize((sqrWidth, sqrWidth))
im_resize = np.array(im_resize)
```

```python
im_resize.shape
```

```python
plt.imshow(im_resize)
```
